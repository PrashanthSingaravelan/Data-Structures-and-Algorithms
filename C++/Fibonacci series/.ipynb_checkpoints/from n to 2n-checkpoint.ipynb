```python
def iter_fib(n):
    sum1 = 0
    t0 = 0
    t1 = 1
    
    if(n<=1):
        return n
    else:
        for i in range(2,2*n):
            sum1 = t0 + t1
            t0 = t1
            t1 = sum1
            
            if i>=n:
                print(sum1, end=' ')

if __name__ == '__main__':
    iter_fib(10)
```
